```python
##TASK 1: Creating Dataframes

import numpy as np
import pandas as pd
```

```python
# Employee DataFrame
employee_data = {
    "ID": ["A001", "A002", "A003", "A004", "A005"],
    "Name": ["John Alter", "Alice Luxumberg", "Tom Sabestine", "Nina Adgra", "Amy Johny"],
    "Gender": ["M", "F", "M", "F", "F"],
    "City": ["Paris", "London", "Berlin", "Newyork", "Madrid"],
    "Age": [25, 27, 29, 31, 30]
}

# Creating of employee dataframe
employee_df = pd.DataFrame(employee_data)

# Saving employee dataframe to CSV
employee_df.to_csv("employee.csv", index=False)
```

```python
# Seniority Level Dataframe

seniority_data = {
    "ID" : ["A001", "A002", "A003", "A004", "A005"],
    "Designation Level" : [2,2,3,2,3]
}

# Creating seniority dataframe
seniority_df = pd.DataFrame(seniority_data)

# Saving seniority dataframe to csv
seniority_df.to_csv("seniority.csv", index=False)

```

```python
# Project DataFrame

project_data = {
    "ID": ["A001", "A002", "A003", "A004", "A005", "A002", "A005", "A003", "A001", "A003", "A001", "A004", "A004", "A005"],
    "Project": ["Project 1", "Project 2", "Project 3", "Project 4", "Project 5", "Project 6", "Project 7", "Project 8", "Project 9", "Project 10", "Project 11", "Project 12", "Project 13", "Project 14"],
    "Cost": [1002000, 2000000, 4500000, 5500000, np.nan, 680000, 400000, 350000, np.nan, 300000, 2000000, 1000000, 3000000, 200000],
    "Status": ["Finished", "Ongoing", "Finished", "Ongoing", "Finished", "Failed", "Finished", "Failed", "Ongoing", "Finished", "Failed", "Ongoing", "Finished", "Finished"]
}

# Creating project dataframe
project_df = pd.DataFrame(project_data)

# Saving project dataframe to CSV
project_df.to_csv("project.csv", index=False)
```

```python
## Task 2: Replacing missing 'Cost' values using running average (loop-based method)

#Load the project dataframe
project_df = pd.read_csv("project.csv")

#initialize running_sum and running_count
running_sum = 0
running_count = 0

# Iterate over the DataFrame using index
for i in range(len(project_df)):
    if pd.isna(project_df.loc[i, 'Cost']):
        #Replacing with running average
        if running_count > 0:
            project_df.loc[i, 'Cost'] = running_sum / running_count
        else:
            project_df.loc[i, 'Cost'] = 0

    else: 
        #update running_sum and running_count
        running_sum += project_df.loc[i,'Cost']
        running_count += 1

#Save the cleaned dataframe
project_df.to_csv("project_cleaned.csv", index=False)

#Display the cleaned dataframe
project_df
```

```python
## Task3: Spliting Employee name coloumn

# Load the employee dataframe
employee_df = pd.read_csv("employee.csv")

#Spliting the 'Name' Column into 'First Name' and 'Last Name'
employee_df[['First_name', 'Last_name']] = employee_df['Name'].str.split(pat=' ', n=1, expand=True)

# Droping the original 'Name' column
employee_df.drop(columns=['Name'], inplace=True)

#Saving the new dataframe to CSV
employee_df.to_csv("employee_name_split.csv", index=False)

#Display the updated dataframe
employee_df
```

```python
 ## Task 4: Merging the three dataframes

# Load the three Dataframes
employee_df = pd.read_csv("employee_name_split.csv")
seniority_df = pd.read_csv("seniority.csv")
project_df = pd.read_csv("project_cleaned.csv")

# Merge employee and seniority on 'ID' and then merge the result again with project data on 'ID'
merged_df = pd.merge(employee_df, seniority_df, on= 'ID', how='left')
merged_df = pd.merge(merged_df, project_df, on= 'ID', how='left')

# Save the Final merged dataframe to csv
merged_df.to_csv("final.csv", index=False)

# Display the result
merged_df
```

```python
## Task 5: Add a new bonus column

# Load the final merged dataframe
merged_df = pd.read_csv("final.csv")

# Create a new 'Bonus' Column - 5% of the cost where project is completed, else 0
merged_df['Bonus'] = merged_df['Status'].apply(
    lambda x: 0 if x != 'Completed' else None
)

# Apply 5% cost bonus only where applicable
merged_df.loc[merged_df['Status'] =='Completed', 'Bonus'] = merged_df['Cost']*0.05

# round off
merged_df['Bonus'] = merged_df['Bonus'].round(2)

# Save the updated dataframe
merged_df.to_csv("final_with_bonus.csv", index=False)

# Display the result
merged_df
```

```python
## Task 6: Demotion and Filtering Based on Project Status and Designation Level

#Load the latest dataframe (with bonus column)
final_df = pd.read_csv("final_with_bonus.csv")

# Step 1: Demote designation level by 1 where project failed
final_df.loc[final_df['Status'].str.lower() == 'fail', 'Designation Level'] -= 1

# Step 2: Remove employess with designation level > 4
final_df = final_df[final_df['Designation Level'] <= 4]

# reset index
final_df = final_df.reset_index(drop=True)

# Save the updated dataframe
final_df.to_csv("final_cleaned.csv", index=False)

#Display the updated data
final_df
```

```python
## Task 7: Add Gender-Based Prefix to First Name and Remove Gender Column

# Load the most recent cleaned dataframe
final_df = pd.read_csv("final_cleaned.csv")

# Add "Mr." or "Mrs." based on Gender to First Name
final_df['First_name'] = final_df.apply(
    lambda row: f"Mr. {row['First_name']}" if row['Gender'] == 'M' else f"Mrs. {row['First_name']}", axis = 1
)
 
# Drop the Gender Column
final_df.drop(columns =['Gender'], inplace=True)

# Save the updated dataframe
final_df.to_csv("Final_named.csv", index=False)

#Display the result
final_df
```

```python
## Task 8: Promote Employees with Age Above 29 by Increasing Designation Level

# Load the latest updated dataframe
final_df = pd.read_csv("final_named.csv")

#Promote designation level by 1 where age > 29
final_df['Designation Level'] = final_df.apply(
    lambda row: row['Designation Level'] + 1 if row['Age'] > 29 else row['Designation Level'], axis = 1
)

# Save the updated dataframe
final_df.to_csv("final_promoted.csv", index=False)

#Display Preview
final_df[['ID', 'First_name', 'Age', 'Designation Level']]
```

```python
## Task 9: Total Project Cost per Employee

# Load the latest updated dataframe
final_df = pd.read_csv("final_promoted.csv")

# Group by ID and First name and sume the cost
total_proj_cost = final_df.groupby(['ID','First_name'])['Cost'].sum().reset_index()

# Rename the 'Cost' column to 'Total Cost'
total_proj_cost.rename(columns = {'Cost': 'Total_cost'}, inplace=True)

# Save to new CSV
total_proj_cost.to_csv("Total_proj_cost.csv", index=False)

# Display Result
total_proj_cost
```

```python
## Task 10: Filter Employees Whose City Name Contains the Letter O

#Load the latest final dataframe
final_df = pd.read_csv("final_promoted.csv")

# Filter employees where City contains the letter 'o' (case insensitive)
city_with_o_df = final_df[final_df['City'].str.contains('o', case=False, na=False)]

# Save to CSV
city_with_o_df.to_csv("city_with_o_csv", index=False)

#Display the Filtered dataframe
city_with_o_df
```
